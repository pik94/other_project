Далее поработаем с пропусками. Попробуем удалить объекты с пропущенными значениями в признаках и количество оставшихся объектов, сравнив его с начальной выборкой.

```python
print('before dropping: %s' % df.shape[0])
print('after dropping: %s' % df.dropna(how='any').shape[0])
```

Ужасно! После удаления объектов с пропущенными данными размер выборки уменьшился более чем в 300 раз. Это говорит о том, что данные пропущенны очень разреженно.

Выведем информацию о том, сколько данных пропущено в каждой колонке, а также общий процент пропусков в конкретной колонке относительно всех данных в данной колонке.

```python
col_missed_info = pd.DataFrame(columns=['feature_name', 'num_miss_data', 
                                        'percent of missed'])

missed_values_count = 0

for i, col in enumerate(df):
    n_miss_values = df.shape[0] - df[col].count()
    percent = round(n_miss_values / df[col].shape[0] * 100, 2)
    
    row = {
        'feature_name': str(i + 1),
        'num_miss_data': n_miss_values, 
        'percent of missed': percent
    }
    col_missed_info = col_missed_info.append(row, ignore_index=True)
    missed_values_count += n_miss_values
```

```python
print('Информация о пропущенных данных по каждой колонке')
col_missed_info
```

```python
size = df.shape[0] * df.shape[1]
print('Процент пропущенных значений среди всех данных: %s%%' % 
      round(missed_values_count / size * 100, 2))
```

Как видно по информацию выше, процент пропущенных данных довольно существеннен. А в одном из признаков и вовсе приближается к 15%.

Посмотрим на гистограммы распредения значений по каждому признаку.

```python
unique_counts = df.nunique()
```

```python
unique_counts[0:5]
```

```python
fig, axes = plt.subplots(nrows=20, ncols=3, figsize=(15, 50))

for i, col in enumerate(df):
    ax = axes[i // 3, i % 3]
    bins = unique_counts[col]
    values = df[col].dropna(how='any')
    if bins < 23:
        ax.hist(values, bins=bins, density=True,
                label='feature %s\nuniques:\n%s' % (i + 1, unique_counts[col]))
    else:
        ax.hist(values, bins=1024, density=True,
                label='feature %s\nuniques:\n%s' % (i + 1, unique_counts[col]))
    ax.set_xlabel('%s' % (i + 1))
    ax.set_ylabel('frequency')
    ax.grid()
    ax.legend()
```

Получили интересную картину. Гистограмма признаков, в которых много уникальных значений, очень напоминает функцию плотности нормального распределения (за исключением признака 37). В остальных случаях картина лишь отдаленно напоминает нормальное, либо не напоминает вовсе (признаки 10, 12, 47).

Рассмотрим более подробно признак 37.
