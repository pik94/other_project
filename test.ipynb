```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print('Shape of X_train: %s, %s' % X_train.shape)
print('Shape of y_train: %s,' % y_train.shape)
print('Shape of X_test: %s, %s' % X_test.shape)
print('Shape of y_test: %s' % y_test.shape)
```

```python
X_train_df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

colors = {0: 'red', 1: 'green', 2: 'blue'}
target = list(map(lambda x: colors.get(x), y_train))
grr = scatter_matrix(X_train_df, c=target, figsize=(15, 15), marker='o')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
predictions
```

```python
print('accuracy: %s' % np.mean(predictions == y_test))
```
